```python
import numpy as np
import tensorflow as tf
```

```python
data_train1 = np.load('train1.npz', allow_pickle=True)
train_input1 = data_train1['inputs'].astype(float)
train_target1 = data_train1['targets'].astype(int)

data_validation1 = np.load('validation1.npz', allow_pickle=True)
validation_input1 = data_validation1['inputs'].astype(float)
validation_target1 = data_validation1['targets'].astype(int)

data_test1 = np.load('test1.npz', allow_pickle=True)
test_input1 = data_test1['inputs'].astype(float)
test_target1 = data_test1['targets'].astype(int)
```

```python
n_samples = len(train_target1)
classes = np.unique(train_target1)
n_classes = len(classes)

classweight1 = {}

for cls in classes:
    n_cls = np.sum(train_target1 == cls)
    classweight1[cls] = n_samples / (n_cls * n_classes)
```

```python
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model1.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)

model1.fit(
    train_input1,
    train_target1,
    batch_size = 100,
    class_weight = classweight1,
    epochs = 100,
    callbacks = early_stop,
    validation_data = (validation_input1, validation_target1),
    verbose = 1
)
```

```python
data_train2 = np.load('train2.npz', allow_pickle=True)
train_input2 = data_train2['inputs'].astype(float)
train_target2 = data_train2['targets'].astype(int)

data_validation2 = np.load('validation2.npz', allow_pickle=True)
validation_input2 = data_validation2['inputs'].astype(float)
validation_target2 = data_validation2['targets'].astype(int)

data_test2 = np.load('test2.npz', allow_pickle=True)
test_input2 = data_test2['inputs'].astype(float)
test_target2 = data_test2['targets'].astype(int)
```

```python
n_samples = len(train_target2)
classes = np.unique(train_target2)
n_classes = len(classes)

classweight2 = {}

for cls in classes:
    n_cls = np.sum(train_target2 == cls)
    classweight2[cls] = n_samples / (n_cls * n_classes)
```

```python
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(64, 'relu'),
    tf.keras.layers.Dense(64, 'relu'),
    tf.keras.layers.Dense(6, 'softmax')
])

model2.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model2.fit(
    train_input2,
    train_target2,
    batch_size = 100,
    class_weight = classweight2,
    epochs = 100,
    callbacks = early_stop,
    validation_data = (validation_input2, validation_target2),
    verbose = 1
)
```

```python
y_pred_prob1 = model1.predict(test_input1)
y_pred1 = (y_pred_prob1 > 0.5).astype(int).flatten()
```

```python
y_pred_prob2 = model2.predict(test_input1)
y_pred2 = tf.argmax(y_pred_prob2, axis=1).numpy()
```

```python
import pandas as pd

test_target22 = pd.read_csv('true_traget_failure.csv')

results_df = pd.DataFrame(test_input1, columns=[
    'Type_H', 'Type_L', 'Type_M', 'Air temperature [K]',
    'Process temperature [K]', 'Rotational speed [rpm]',
    'Torque [Nm]', 'Tool wear [min]'
])

results_df['True_Binary'] = test_target1
results_df['Predicted_Failure_Binary'] = y_pred1
results_df['Predicted_Failure_Type'] = y_pred2
results_df['True_Failure'] = test_target22
results_df.loc[results_df['Predicted_Failure_Binary'] < 0.5, 'Predicted_Failure_Type'] = 0
```

```python
results_df['Predicted_Failure_Type'] = results_df['Predicted_Failure_Type'].map({
     0:'No Failure',
     1:'Power Failure',
     2:'Heat Dissipation Failure', 
     3:'Overstrain Failure',
     4:'Tool Wear Failure',
     5:'Random Failures'
})
```

```python
results_df.to_csv('Output.csv', index=False)
```
